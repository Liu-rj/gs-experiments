```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_names = os.listdir('graphsage/outputs/data/')
file_names
```

```python
dfs = {}
for name in file_names:
    key = name.split('_')[1]
    dfs[key] = pd.read_csv('graphsage/outputs/data/' + name)
```

```python
plt.figure(figsize=(15, 8))
keys = sorted(dfs.keys())
for key in keys:
    if key == 'pyg':
        continue
    time = dfs[key]['cumulated time/s']
    acc = dfs[key]['acc']
    smoothed = acc.copy()
    for i, v in acc.items():
        if i > 2:
            smoothed[i] = acc[i-2:i+2].mean()
    plt.plot(time, smoothed, label=key)
plt.grid(alpha=0.2)
plt.ylabel('Validation Accuracy')
plt.xlabel('Time/s')
plt.title('Time To Accuracy')
# plt.ylim(0.8, 0.92)
# plt.xlim((-5, 1000))
plt.legend()
plt.savefig('graphsage_time2acc.eps', format='eps', bbox_inches='tight')
plt.show()
```
