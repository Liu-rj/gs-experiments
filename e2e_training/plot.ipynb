```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from brokenaxes import brokenaxes
```

```python
file_names = os.listdir('graphsage/outputs/data/')
file_names
```

```python
dfs = {}
for name in file_names:
    path = 'graphsage/outputs/data/' + name
    if os.path.isdir(path):
        continue
    key = name.split('_')[1]
    dfs[key] = pd.read_csv(path)
```

```python
plt.figure(figsize=(15, 8))
ax = brokenaxes(xlims=((0, 500), (13000, 13100)), despine=False, wspace=0.1, d=0.01)
keys = sorted(dfs.keys())
max_acc = 1
lines = []
for key in keys:
    time = dfs[key]['cumulated time/s']
    acc = dfs[key]['acc']
    smoothed = acc.copy()
    max_acc = min(max(smoothed), max_acc)
    lines.append((time, smoothed, key))

for line in lines:
    time, acc, key = line
    ax.plot(time, acc, label=key)
ax.axhline(y=max_acc, color='r', linestyle='--')
ax.grid(alpha=0.2)
plt.ylabel('Validation Accuracy')
plt.xlabel('Time/s')
plt.title('Time To Accuracy')
# plt.ylim(0.8, 0.92)
# plt.xlim((-10, 2000))
ax.legend(loc='lower right')
plt.savefig('graphsage_time2acc.eps', format='eps', bbox_inches='tight')
plt.show()
```

```python
sliding_window = 2
plt.figure(figsize=(15, 8))
ax = brokenaxes(xlims=((0, 500), (3300, 3400)), despine=False, wspace=0.1, d=0.01)
keys = sorted(dfs.keys())
max_acc = 1
lines = []
for key in keys:
    time = dfs[key]['cumulated time/s']
    acc = dfs[key]['acc']
    smoothed = acc.copy()
    for i, v in acc.items():
        left = i - sliding_window
        if left < 0:
            left = 0
        right = i + sliding_window
        if right > len(acc):
            right = len(acc)
        smoothed[i] = acc[left:right].mean()
    # for i, v in acc.items():
    #     if i > 2:
    #         smoothed[i] = acc[i-2:i+2].mean()
    max_acc = min(max(smoothed), max_acc)
    lines.append((time, smoothed, key))

for line in lines:
    time, acc, key = line
    acc = acc.to_numpy()
    index = np.argwhere(acc>=max_acc)[0][0] + 1
    ax.plot(time[:index], acc[:index], label=key)
ax.axhline(y=max_acc, color='r', linestyle='--')
ax.grid(alpha=0.2)
ax.set_xlabel('Time/s')
ax.set_ylabel('Validation Accuracy')
ax.set_title('Time to Accuracy')
ax.legend(loc='lower right')
plt.savefig('graphsage_time2acc.eps', format='eps', bbox_inches='tight')
plt.show()
```
